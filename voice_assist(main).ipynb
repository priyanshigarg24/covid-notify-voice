```python
TotalcasesIndex=0
NewcasesIndex=1
CasespermillionpeopleIndex=3
deathsIndex=4
data={}
def getLocationData(location):
    html=requests.get("https://news.google.com/covid19/map?hl=en-IN&mid=%2Fm%2F03rk0&gl=IN&ceid=IN%3Aen")
    soup = BeautifulSoup(html.text, 'html.parser')
    tab = soup.find("table",{"class":"pH8O4c"})
    rows=tab.find_all('tr')
    data={}
    for tr in rows:
        cols = tr.find_all('td')
        heading=tr.find('th')
        rowHead=heading.text.strip().lower() #The strip() method removes any leading (spaces at the beginning) and trailing (spaces at the end) characters (space is the default leading character to remove)


        cols = [ele.text.strip() for ele in cols]
        # print(rowHead)
        # print(cols)
        data[rowHead]=cols
    return data.get(location)
# print(data)
```

```python
def getVaccineData(location):
    html1=requests.get("https://news.google.com/covid19/map?hl=en-IN&mid=%2Fm%2F03rk0&gl=IN&ceid=IN%3Aen&state=4")
    soup1 = BeautifulSoup(html1.text, 'html.parser')
    tab1 = soup1.find("table",{"class":"pH8O4c"})
    rows1=tab1.find_all('tr')
    data1={}
    for tr1 in rows1:
        cols1 = tr1.find_all('td')
        heading1=tr1.find('th')
        rowHead1=heading1.text.strip().lower()
        cols1 = [ele.text.strip() for ele in cols1]
        # print(rowHead)
        # print(cols)
        data1[rowHead1]=cols1
    return data1.get(location)
```
